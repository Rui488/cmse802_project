```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

**Load the dataset**

```python
df = pd.read_csv(r"C:\Users\白猪白猪\Documents\Documents\CMSE802\Class_Material\Homework\Project\Data\imputed_pca_data.csv")
# Isolate the numerical columns from the dataset for further analysis
numerical_df = df.iloc[:, 3:]
```

**Plot histogram**

```python
plt.figure(figsize=(30, 20))
numerical_df.hist(bins=50)
plt.title('Histogram of Features')
plt.tight_layout()
plt.show()
```

**Plot pairplot**

```python
# Calculate the Spearman correlation coefficient matrix
corr_matrix = numerical_df.corr(method='spearman')

# Plot the correlation matrix
g = sns.pairplot(numerical_df, kind="scatter", diag_kind="kde")

# Calculate Spearman correlation coefficients and display them on each scatter plot
for i in range(len(g.axes)):
    for j in range(len(g.axes)):
        if i < j:  # Only calculate and display correlations for the upper triangle to avoid duplicates
            corr_value = corr_matrix.iloc[i, j]
            g.axes[j][i].text(0.95, 0.85, f'{corr_value:.2f}',
                               horizontalalignment='right',
                               verticalalignment='top',
                               transform=g.axes[j][i].transAxes,
                               color='black', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))


# Adjust layout to accommodate the text
plt.tight_layout()

# Display the plot
plt.show()
```
